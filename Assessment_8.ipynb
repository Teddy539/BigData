```
cutter <- worker(bylines = F)
```

```
cutter[content]
```

```
new_words <- c("美國總統大選","選舉人票","6票","20票","15票","16票","3票","11票","亞歷桑那州","福斯新聞","5日","上午9時","沒有計算在內","州內","4州","任何一州","88選票","11張票","搖擺州","7萬票","威斯康辛州","重新計票","選舉人團成員","12月中旬","副總統","隔年","1月初","正副總統","2021年1月20日","45州","214張","264張","5州","分別是","214張","好幾個郡","88選票","20萬張","郵寄選票","538張","270張","州治集會","略為","民主黨據點","這州","競選團隊","提起訴訟","美國總統","新任總統","形式上","下一屆","當地時間","才會","大勢底定","目前為止","川普陣營")
writeLines(new_words,"new_words.txt")

stop_words <- c("已","於","至","的","但","而","也","仍","就","更","在","有","開","從","以")
writeLines(stop_words,"stop_words.txt")

cutter <- worker(user = "new_words.txt",stop_word = "stop_words.txt",bylines = FALSE)
seg_words <- cutter[content]
seg_words
```

```
txt_freq <- freq(seg_words)
txt_freq <- arrange(txt_freq, desc(freq))
head(txt_freq)
```
