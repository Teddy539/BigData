```
sick <- read.csv("sick.csv")
sick
```

討論一:台灣登革熱在**2003年到2020年的發病次數**統計

```
sick %>%
 count(發病年份) %>%
 ggplot(aes(x=發病年份 ,y=n,fill = 發病年份)) +
 geom_col() +
 geom_text(aes(label = n),vjust = 0) +
 labs(title = "不同年分的發病次數",x = "發病年份",y = "數量") + theme(legend.position= "none") #用長條圖表示各年分及人數的對照
```

討論二:台南市各地區確診"**性別**"及"**是否為境外移入**"人數統計

```
sick %>%
 dplyr::filter(縣市 == "台南市") %>%
 dplyr::select(鄉鎮,性別,是否為境外移入) #列出台南市發病者的性別及是否為境外移入
```

```
sick %>%
 count(性別) %>%
 ggplot(aes(x = 性別,y = n)) +
 geom_col() #以長條圖表示男女個別的發病人數
```

```
sick %>%
 count(是否為境外移入) %>%
 ggplot(aes(x = 是否為境外移入,y = n)) +
 geom_col() #以長條圖表示這些發病者是否為境外移入
```

討論三:發病高峰的**月份**為何?

```
sick %>%
 count(發病月份) %>%
 ggplot(aes(x=發病月份,y=n,fill = 發病月份)) +
 geom_col() +
 labs(title = "不同月分的發病次數",x = "發病月份",y = "數量") #以長條圖表示各月份發病人數，以找出發病的高峰期
```

討論四:各**年齡層**的**發病次數**統計

```
sick %>%
 count(年齡層) %>%
 ggplot(aes(x=年齡層,y=n,fill = 年齡層)) +
 geom_col() +
 labs(title = "各個年齡層的發病人數",x = "年齡層",y = "數量") #以長條圖表示出各個年齡層的發病人數
```
