```
dataset<-data.frame(gp=factor(rep(letters[1:3],each=10)),y=rnorm(30))
```

```
dataset$z<-dataset$y+rnorm(30)
```

```
ggplot(dataset,aes(x=z,y=y))+geom_point()+facet_grid(gp~.)+geom_smooth()
ggplot(dataset,aes(x=z,y=y))+geom_point()+facet_grid(.~gp)+geom_smooth()
```

```
ggplot(dataset,aes(x=z,y=y))+geom_line()+facet_grid(gp~.)
ggplot(dataset,aes(x=z,y=y,color=gp))+geom_line()
```
