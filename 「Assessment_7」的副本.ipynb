```
ggplot(rays,aes(x=PublishAgency,y=County))+geom_point()
ggplot(rays,aes(x=UVI,y=SiteName))+geom_point()
```

```
ggplot(rays,aes(x=SiteName))+geom_bar()
ggplot(rays,aes(x=County))+geom_bar()
```
