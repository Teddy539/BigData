```
install.packages("SportsAnalytics") #安裝SportsAnalytics
```

```
library(SportsAnalytics) #讀取SportsAnalytics
library(ggplot2) #讀取ggplot2
```

```
NBA1213 <- fetch_NBAPlayerStatistics("12-13") #將NBA1213設為data frame
```

```
str(NBA1213) #瀏覽data
```

```
NBA1213Order <- NBA1213[order(NBA1213$FieldGoalsMade,decreasing = T),] #將FieldGoalsMade由大到小排序
NBA1213Order 
```

```
subset(NBA1213,Position=="C") #用subset()搜尋Position中的C位置
```

```
NBA1213[grepl("NYK",NBA1213$Team),] #用grepl()搜尋Team中的NYK隊伍
```

```
ggplot(NBA1213,aes(x=Blocks,y=Steals))+geom_point() #用散佈圖表示Blocks和Steals的關係，X軸為Blocks，Y軸為Steals
```

```
ggplot(NBA1213,aes(x=Technicals))+geom_bar() #用長條圖表示Technicals的次數
```

```
ggplot(NBA1213,aes(x=GamesPlayed,y=Steals))+geom_line() #用折線圖表示GamesPlayed和Steals的關係，X軸為GamesPlayed，Y軸為Steals
```

```
ggplot(NBA1213,aes(x=FieldGoalsAttempted,y=Team))+geom_boxplot() #用合狀圖表示FieldGoalsAttempted和Team的關係，X軸為FieldGoalsAttempted，Y軸為Team
```
